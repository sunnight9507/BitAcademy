```python
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
chrome_driver = 'chromedriver'
driver = webdriver.Chrome(chrome_driver)

url = 'http://itempage3.auction.co.kr/detailview.aspx?ItemNo=B705362500&listqs=class%3dCorner.CategoryBest%26listorder%3d1&listtitle=%ba%a3%bd%ba%c6%ae100&frm2=through#vip_tab_comment'
# chrome driver 로 해당 페이지가 물리적으로 open
driver.get(url)

src = driver.page_source # page source 가져오기
soup = BeautifulSoup(src)

# chrome driver 사용 후, close 함수로 종료
driver.close()

for strings in soup.find_all('div', class_ = "box__content"):
    print([x for x in [i.strip() for i in strings.get_text().split('\n')] if x])

    # 평점
    print('평점 : {}'.format(int(int(strings.find('span', class_ = 'sprite__vip image__star-fill')['style'][-4:-1]) / 10)))
    print()
    print()
    print()
```
